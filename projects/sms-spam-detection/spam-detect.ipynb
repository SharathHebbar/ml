```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB


from sklearn.metrics import accuracy_score

from sklearn.feature_extraction.text import CountVectorizer

import pickle
```

```python
data = pd.read_csv('spam.csv',encoding= 'latin-1')
data
```

```python
data = data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis = 1)
data
```

```python
data = data.rename(columns={'class': 'result'})
```

```python
data['spam'] = data.result.map({'spam':1,'ham':0}) 
#data.result
```

```python
Y = np.array(data['result'])
X = np.array(data['message'])
X,Y
```

```python
cv = CountVectorizer()
```

```python
x = cv.fit_transform(X)
y = (Y)

xtrain,xtest,ytrain,ytest = train_test_split(x,y)
clf = MultinomialNB()
#clf = SVC()
clf.fit(xtrain,ytrain)
```

```python
yp=clf.predict(xtest)
acc = accuracy_score(ytest,yp)
acc
```

```python
clf.score(xtest,ytest)
```

```python
yp=clf.predict(xtest)
for i in range(len(yp)):
    if yp[i] == ytest[i]:
        print("--same--",yp[i],ytest[i])
    else:
        print("---diff----",yp[i],ytest[i])
```

```python
sample = input('Enter a message:')
res = cv.transform([sample]).toarray()
print(clf.predict(res))
```

```python
x = cv.fit_transform(X)
y = (Y)


best = 0
bestmodel = None
for i in range(1000):
    xtrain,xtest,ytrain,ytest = train_test_split(x,y)
    clf = MultinomialNB()
    clf.fit(xtrain,ytrain)
    yp=clf.predict(xtest)
    acc = accuracy_score(ytest,yp)
    if(acc > best):
        best = acc
        with open("spam.pickle","wb") as f:
            pickle.dump(clf,f)
best
```

```python
bestmodel = open('spam.pickle','rb')
clf = pickle.load(bestmodel)
clf.predict(xtest)
yp=clf.predict(xtest)
acc = accuracy_score(ytest,yp)
clf.score(xtest,ytest),acc
```
