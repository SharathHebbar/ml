### Color space is HSV
Infact HSV is very useful in color filtering.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('road.jpg')
image
```

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
hsv_image
```

```python
r,c = 2,2
fig, axis = plt.subplots(r, c, figsize=(15, 10))
fig.tight_layout()

axis[0][0].imshow(cv2.cvtColor(hsv_image, cv2.COLOR_BGR2RGB))
axis[0][0].set_title("HSV Image")

axis[0][1].imshow(cv2.cvtColor(hsv_image[:,:,0], cv2.COLOR_BGR2RGB))
axis[0][1].set_title("Hue Image")

axis[1][0].imshow(cv2.cvtColor(hsv_image[:,:,1], cv2.COLOR_BGR2RGB))
axis[1][0].set_title("Saturation Image")

axis[1][1].imshow(cv2.cvtColor(hsv_image[:,:,2], cv2.COLOR_BGR2RGB))
axis[1][1].set_title("Value Image")

plt.show()
```

```python
cv2.imwrite('hsvimage.jpg',hsv_image)
cv2.imwrite('hsvimage.png',hsv_image)
```

```python
cv2.imwrite('hueimage.jpg',hsv_image[:,:,0])
cv2.imwrite('hueimage.png',hsv_image[:,:,0])
```

```python
cv2.imwrite('saturation_image.jpg',hsv_image[:,:,1])
cv2.imwrite('saturation_image.png',hsv_image[:,:,1])
```

```python
cv2.imwrite('value_image.jpg',hsv_image[:,:,2])
cv2.imwrite('value_image.png',hsv_image[:,:,2])
```
