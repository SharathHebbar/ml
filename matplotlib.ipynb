```python
x = [2,5,6,7,8]
y = [5,6,7,8,9]

x2 =[4,5,7,8,9]
y2 = [3,5,7,8,9]
```

```python
plt.plot(x,y, color = 'g', linewidth=6)
plt.plot(x2,y2, color = 'r', linewidth=4)

plt.title('Comparison graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.annotate(xy=[2,5.5], text= "First Line")
plt.annotate(xy=[3,3], text= "Second Line")

plt.legend(['First_line', 'Second_Line'],loc = 4 )

plt.show()
```

# Bar Graph

# Vertical Bar Graph


```python
x = [2,5,6,7,8]
y = [5,6,7,8,9]

x2 =[4,5,7,8,9]
y2 = [3,5,7,8,9]
```

```python
plt.bar(x,y, color = 'g', linewidth=6)
plt.bar(x2,y2, color = 'r', linewidth=4)

plt.title('Comparison graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.annotate(xy=[2,5.5], text= "First Line")
plt.annotate(xy=[3,3], text= "Second Line")

plt.legend(['First_line', 'Second_Line'],loc = 4 )

plt.show()
```

# Horizontal Bar Graph

```python
plt.barh(x,y, color = 'g', linewidth=6)
plt.barh(x2,y2, color = 'r', linewidth=4)

plt.title('Comparison graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.legend(['First_line', 'Second_Line'],loc = 4 )

plt.show()
```
