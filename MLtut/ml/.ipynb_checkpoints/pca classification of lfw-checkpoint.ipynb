```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report, confusion_matrix

from sklearn import ensemble
from time import time
```

```python
lfw = datasets.fetch_lfw_people(min_faces_per_person = 100, resize=0.4)
```

```python
lfw.keys()
```

```python
lfw.data.shape
```

```python
lfw.images.shape
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8 ,i+1)
    ax.imshow(lfw.images[i], cmap=plt.cm.bone)
plt.show()
```

```python
x,y = lfw.data, lfw.target


x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)



pcao = PCA()
pcao.fit(x_train)
```

```python
k = 0
total= sum(pcao.explained_variance_)
currsum = 0
while currsum/total < 0.98:
    currsum += pca.explained_variance_[k] 
    k=k+1
k
```

```python

pcal = PCA(n_components=k, whiten= True)
t = pcal.fit_transform(x_train)
pcal.fit(x)
t.shape
```

```python
xa = pcal.inverse_transform(t).reshape(855,50,37)
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8 ,i+1)
    ax.imshow(xa[i], cmap=plt.cm.bone)
plt.show()
```

```python
pcao.components_.shape
```

```python
pcal.components_.shape
```

```python
eigen = pcal.components_.reshape(k,50,37)
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8 ,i+1)
    ax.imshow(eigen[i], cmap=plt.cm.bone)
plt.show()
```

```python
xtrainpca = t
xtestpca = pcal.transform(x_test)
```

## Random Forest

#### without applying PCA

```python
rf = ensemble.RandomForestClassifier()
s = time()
rf.fit(x_train,y_train)
print("Time: ",time()-s)
y_pred = rf.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print("Accuracy: ",rf.score(x_test,y_test))
```

#### with applying PCA

```python
rf = ensemble.RandomForestClassifier()
s = time()
rf.fit(xtrainpca,y_train)
print("Time: ",time()-s)
y_pred = rf.predict(xtestpca)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print("Accuracy",rf.score(xtestpca,y_test))
```
