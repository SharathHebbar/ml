### Importing libraries

```python
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

import seaborn as sns
```

#### Loading csv files

```python
data = pd.read_csv('train.csv')
data1 = pd.read_csv('test.csv')
data
```

### Data Preprocessing

```python
data = data.drop(['Name','Cabin','Ticket'],axis=1)
data1 = data1.drop(['Name','Cabin','Ticket'],axis=1)
print(data.head())
print(data1.head())
```

```python
le = LabelEncoder()
```

```python
a = le.fit_transform(list(data["Pclass"]))
b = le.fit_transform(list(data["Sex"]))
c = le.fit_transform(list(data["Age"]))
d = le.fit_transform(list(data["SibSp"]))
e = le.fit_transform(list(data["Parch"]))
f = le.fit_transform(list(data["Fare"]))
g = le.fit_transform(list(data["Embarked"]))
h = le.fit_transform(list(data["Survived"]))
```

```python
x_train = list(zip(a,b,c,d,e,f,g))
y_train = list(h)
```

```python
a = le.fit_transform(list(data1["Pclass"]))
b = le.fit_transform(list(data1["Sex"]))
c = le.fit_transform(list(data1["Age"]))
d = le.fit_transform(list(data1["SibSp"]))
e = le.fit_transform(list(data1["Parch"]))
f = le.fit_transform(list(data1["Fare"]))
g = le.fit_transform(list(data1["Embarked"]))
x_test = list(zip(a,b,c,d,e,f,g))
x_test
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
z = y_pred
print("Pclass,Sex,Age,SibSp,Parch,Fare,Embarked,Survived")
for i in range(len(z)):
    print(x_test[i],z[i])
```
