```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(2343243)
```

```python
mean_v1 = np.array([0,0,0])
cov_m1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
c1 = np.random.multivariate_normal(mean_v1,cov_m1,100)
```

```python
mean_v2 = np.array([1,1,1])
cov_m2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
c2 = np.random.multivariate_normal(mean_v2,cov_m2,100)
```

```python
from mpl_toolkits.mplot3d import Axes3D, proj3d
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111,projection='3d')
ax.plot(c1[:,0],c1[:,1],c1[:,2],'o')
ax.plot(c2[:,0],c2[:,1],c2[:,2],'^')
plt.show()
```

```python
all_d = np.concatenate((c1,c2))
all_d
```

```python
pca = PCA(n_components= 2)
t = pca.fit_transform(all_d)
t
```

```python
pca.components_
```

```python
plt.plot(t[0:100,0] ,t[0:100,1],"o" )
plt.plot(t[100:200,0] ,t[100:200,1],"^" )
plt.show()
```

```python
xa = pca.inverse_transform(t)
xa
```

```python
from mpl_toolkits.mplot3d import Axes3D, proj3d
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111,projection='3d')
ax.plot(xa[0:100,0],xa[0:100,1],xa[0:100,2],'o')
ax.plot(xa[100:200,0],xa[100:200,1],xa[100:200,2],'^')
plt.show()
```

```python
all_d = all_d.T
cov = np.cov((all_d))
print(cov.shape)
print(cov)
```

```python
eigen_val, eigen_vectors =np.linalg.eig(cov)
eigen_val, eigen_vectors
```

```python
pca.components_
```

```python
eig_val_vector_pair = []
for i in range(len(eigen_val)):
    eig_vec = eigen_vectors[:,i]
    eig_val_vector_pair.append((eigen_val[i],eig_vec))
eig_val_vector_pair.sort(reverse=True)
eig_val_vector_pair
```
