```python
data = data.drop(['PassengerId','Ticket','Name'],axis=1)
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
a = le.fit_transform(list(data["Pclass"]))
b = le.fit_transform(list(data["Sex"]))
c = le.fit_transform(list(data["Age"]))
d = le.fit_transform(list(data["SibSp"]))
e = le.fit_transform(list(data["Parch"]))
f = le.fit_transform(list(data["Fare"]))
g = le.fit_transform(list(data["Embarked"]))
h = le.fit_transform(list(data["Survived"]))
```

```python
x = list(zip(a,b,c,d,e,f,g))
y = list(h)
```

```python
from sklearn.model_selection import train_test_split as tt
x_train,x_test,y_train,y_test = tt(x,y)
```

```python
dtc = DecisionTreeClassifier(random_state=0)
dtc.fit(x_train,y_train)
y_predict = dtc.predict(x_test)
dtc.score(x_train,y_train),dtc.score(x_test,y_test)
```

```python
rfc = RandomForestClassifier(max_depth = 17, random_state=0)
rfc.fit(x_train,y_train)
rfc.score(x_train,y_train),rfc.score(x_test,y_test)
```
