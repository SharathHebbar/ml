## Approach 4 imputer

#### Startegy to handle missing values
    mean
    median
    most_frequent

```python
#from sklearn.preprocessing import Imputer
#renamed as
from sklearn.impute import SimpleImputer
```

```python
imputer_mean = SimpleImputer(missing_values="NaN", strategy='mean')
imputer_median = SimpleImputer(missing_values="NaN", strategy='median')
print(imputer_mean,imputer_median)
```

```python
imputer_mean_data = imputer_mean.fit_transform(data[cols])
imputer_median_data = imputer_median.fit_transform(data[cols])
```
