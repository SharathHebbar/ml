### Importing libraries

```python
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

### Loading data

```python
data = pd.read_csv("car.csv")
data.head()
```

#### Label Encoding

```python

l= LabelEncoder()
#l = preprocessing.LabelEncoder()
```

```python
buying = l.fit_transform(list(data["buying"]))
maint = l.fit_transform(list(data["maint"]))
door = l.fit_transform(list(data["door"]))
persons = l.fit_transform(list(data["persons"]))
lug_boot = l.fit_transform(list(data["lug_boot"]))
safety = l.fit_transform(list(data["safety"]))
cls = l.fit_transform(list(data["class"]))
door
```

##### zip fun to combine all the data


```python
#features
x = list(zip(buying,maint,door,persons,lug_boot,safety))
#labels
y = list(cls)
```

#### Training and testing data 

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

#### Model training

```python
model = KNeighborsClassifier(n_neighbors=9)
model.fit(x_train,y_train)
```

##### accuracy

```python
acc= model.score(x_test,y_test)
acc
```

### predictions

```python
pred = model.predict(x_test)
names = ["unacc","acc","good","vgood"]
for i in range(len(pred)):
    print("Predicted: ",names[pred[i]],"Data: ",x_test[i],"Actual: ",names[y_test[i]])
```
